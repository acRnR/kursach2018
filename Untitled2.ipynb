```python
def clear_nan(lll):
    for www in lll:
        if type(www) == float:
            lll.remove(www)
            #print('kek')
            break
    #print(lll)
    return lll
```

```python
lingua = clear_nan(list(set(df['Язык'])))
russo = clear_nan(list(set(df['Русский язык'])))
livello = clear_nan(list(set(df['Уровень'])))
gruppo = clear_nan(list(set(df['группа'])))
ordine = clear_nan(list(set(df['расположение частей'])))
principale = clear_nan(list(set(df['подлеж гл'])))
inferiore = clear_nan(list(set(df['подлеж прид'])))
```

```python
def write_data(namae, numero):
    fl = open('rlc_data.csv', 'a', encoding='UTF-8')
    fl.write(namae + ';' + numero + '\n')
    fl.close
```

```python
#all_calc = {}
def all_calculations(lang, rus, lev, cg, ordr, mc, sc):
    nome = lang  + ';' + rus  + ';' + lev + ';' + cg + ';' + ordr + ';' + mc + ';' + sc
    #print (nome, type(nome))
    nombre = ((df['Язык']==lang)&(df['Русский язык']==rus)&(df['Уровень']==lev)&(df['группа']==cg)&(df['расположение частей']==ordr)&(df['подлеж гл']==mc)&(df['подлеж прид']==sc)).sum()
    write_data(nome, str(nombre))
    #all_calc[nome] = nombre
    print('..', end=' ')
```

```python
for la in lingua:
    for ru in russo:
        for le in livello:
            for ccg in gruppo:
                for orrd in ordine:
                    for mmc in principale:
                        for ssc in inferiore:
                            all_calculations(la, ru, le, ccg, orrd, mmc, ssc)
```
